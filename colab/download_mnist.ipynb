```python
import tensorflow as tf
import numpy as np
import os

# Load the MNIST dataset
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.mnist.load_data()

# Define the path to save the dataset in Google Drive
save_path = '/content/drive/MyDrive/mnist_dataset.npz'

# Create the directory if it doesn't exist
os.makedirs(os.path.dirname(save_path), exist_ok=True)

# Save the dataset to Google Drive
np.savez_compressed(save_path, x_train=x_train, y_train=y_train, x_test=x_test, y_test=y_test)

print(f"MNIST dataset saved to {save_path}")
```

```python
import numpy as np

# Define the path to the saved dataset in Google Drive
save_path = '/content/drive/MyDrive/mnist_dataset.npz'

# Load the dataset from Google Drive
with np.load(save_path) as data:
    x_train = data['x_train']
    y_train = data['y_train']
    x_test = data['x_test']
    y_test = data['y_test']

print("MNIST dataset loaded successfully.")
print(f"x_train shape: {x_train.shape}")
print(f"y_train shape: {y_train.shape}")
print(f"x_test shape: {x_test.shape}")
print(f"y_test shape: {y_test.shape}")
```

```python
import matplotlib.pyplot as plt

# Print one of the loaded arrays (e.g., x_train)
print("Displaying x_train[0] as an image:")
plt.imshow(x_train[0], cmap='gray')
plt.title(f"Label: {y_train[0]}")
plt.axis('off')  # Hide axes
plt.show()
```
